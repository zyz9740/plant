```python
import os
import cv2 
import PIL
from PIL import Image
import scipy.io
import numpy as np
data = scipy.io.loadmat('imagelabels.mat')
from matplotlib import pyplot as plt
```

```python
#import string
data_labels = data['labels'][0]
print(data_labels[0])
a = data_labels[0]
path = './data_plants/' + 'train/' + str(a)
print(path)
```

```python
def process(img):
    rows, cols, channel = img.shape
    t_mat = np.array([[1, 0, 20], [0, 1, 10]], np.float32)
    t_mm = cv2.getRotationMatrix2D((cols/2, rows/2), 30, 1)
    '''t_mat.at<float>(0, 0) = 1;
    t_mat.at<float>(0, 2) = 20; //水平平移量
    t_mat.at<float>(1, 1) = 1;
    t_mat.at<float>(1, 2) = 10; //竖直平移量'''
    return cv2.warpAffine(img, t_mat, (rows, cols))
```

```python
i = 0
k = 0
old_p = ''
for filename in os.listdir(r"./102flowers/jpg"):
    img = Image.open("./102flowers/jpg/"+filename)
    im = np.array(img)#cv2.imread("./102flowers/jpg/"+filename, cv2.IMREAD_COLOR)#Image.open("./102flowers/jpg/"+filename)
    #img = img.resize((80, 80), PIL.Image.ANTIALIAS)
    path = './data_plants_org/' + 'train/' + str(data_labels[i])+ '/'
    i += 1
    isExists=os.path.exists(path)
    if old_p != path:
        k = 5
        old_p = path
    if k:
        isExists=os.path.exists(path+'0_'+filename)
        if isExists:
            os.remove(path+'0_'+filename)
        path = './data_plants_org/' + 'val/' + str(data_labels[i])+ '/'
        isExists=os.path.exists(path)
        if not isExists:
            os.makedirs(path) 
        k -= 1
        
        img_1 = process(im)
        #img_1 = img_1[:,:,0]
        img=Image.fromarray(img_1)
        img.save(path+'0_'+filename)
    else:
        img_1 = process(im)
        #img_1 = img_1[:,:,0]
        img=Image.fromarray(img_1)
        img.save(path+'0_'+filename)
        path = './data_plants_org/' + 'val/' + str(data_labels[i])+ '/'
        isExists=os.path.exists(path+'0_'+filename)
        if isExists:
            os.remove(path+'0_'+filename)
```
